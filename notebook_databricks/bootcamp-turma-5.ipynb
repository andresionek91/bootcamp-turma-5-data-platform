# Streaming

```
#spark.read.table('glue_data_lake_raw.atomic_events') carregando a tabela do glue

data_schema = spark.read.format('json').load('s3://s3-belisquito-turma-5-production-data-lake-raw/atomic_events/').schema

df = spark.readStream.format('json').schema(data_schema).load('s3://s3-belisquito-turma-5-production-data-lake-raw/atomic_events/')

display(df)
```

```
from pyspark.sql import functions as F

df = df.withColumn('etl_timestamp', F.current_timestamp())
df = df.withColumn('event_date', F.to_date(F.col('event_timestamp')))
df = df.withColumn('source_filename', F.input_file_name())

display(df)
```

```
df.createOrReplaceTempView('atomic_events_stream')
```

```
%sql SELECT source_filename, count(1) from atomic_events_stream group by 1
```

```
df.writeStream.format('delta').partitionBy('event_date').option("checkpointLocation", "dbfs:/checkpoints/atomic_events").start('s3://s3-belisquito-turma-5-production-data-lake-staged/atomic_events/')
```

# Merge

```
data_schema = spark.read.format('parquet').load('s3://s3-belisquito-turma-5-production-data-lake-raw/orders/public/orders_v2').schema

df = spark.read.format('parquet').schema(data_schema).load('s3://s3-belisquito-turma-5-production-data-lake-raw/orders/public/orders_v2')
df = df.withColumn('etl_timestamp', F.current_timestamp())
df = df.withColumn('created_date', F.to_date(F.col('created_at')))
df = df.withColumn('source_filename', F.input_file_name())

df.limit(1).write.format('delta').partitionBy('created_date').save('s3://s3-belisquito-turma-5-production-data-lake-staged/orders')

df = spark.readStream.format('parquet').schema(data_schema).load('s3://s3-belisquito-turma-5-production-data-lake-raw/orders/public/orders_v2')
```

```
from delta.tables import DeltaTable
from pyspark.sql.window import Window

from pyspark.sql import functions as F

deltaTable = DeltaTable.forPath(spark, path='s3://s3-belisquito-turma-5-production-data-lake-staged/orders')


def _upsert(microbatch, batch_id):
    microbatch = microbatch.withColumn('row_rank', F.row_number().over(Window.partitionBy(F.col('extracted_at')).orderBy(
                F.col("extracted_at").desc())))
    microbatch = microbatch.filter(F.col('row_rank').isin(1))
    microbatch = microbatch.drop("row_rank")
    
    deltaTable.alias("target").merge(microbatch.alias('updates'), "target.order_id = updates.order_id").whenMatchedUpdateAll("updates.Op='U'").whenNotMatchedInsertAll().execute()
  

df = df.withColumn('etl_timestamp', F.current_timestamp())
df = df.withColumn('created_date', F.to_date(F.col('created_at')))
df = df.withColumn('source_filename', F.input_file_name())

df.writeStream.format('delta').option("checkpointLocation", "dbfs:/checkpoints/orderv_v2").outputMode('update').foreachBatch(_upsert).start()
```

```
df_final = spark.read.format('delta').load('s3://s3-belisquito-turma-5-production-data-lake-staged/orders')
df_final.createOrReplaceTempView('orders')
```
